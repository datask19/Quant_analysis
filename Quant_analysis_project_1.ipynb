```python
# Importing necessary libraries (numpy, pandas, matplotlib).
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Creating a DataFrame of random returns and cumulating them to generate prices.
returns = pd.DataFrame(np.random.normal(1.0, 0.03, (100, 10)))
prices = returns.cumprod()

# Plotting the generated prices with titles, labels, and legend.
prices.plot(figsize=(14, 7))
plt.title('Randomly generated prices')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend(loc=0);
```

```python
# Assuming your timestamps are in UTC (you may need to adjust this to your actual timezone)
prices = prices.tz_localize('UTC')

# Now, you can convert the timezone to 'US/Eastern'
eastern_prices = prices.tz_convert('US/Eastern')
eastern_prices.head(10)
```

```python
# Defining a custom resampler to return the first value of each month and applying it to the stock prices.
def custom_resampler(array_like):
    """ Returns the first value of the period """
    return array_like.iloc[0]

first_of_month_prices = prices.resample('M').agg(custom_resampler)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create data
x = np.linspace(0, 2 * np.pi, 100)  # X values from 0 to 2*pi
y1 = np.sin(x)  # First function (sine)
y2 = np.cos(x)  # Second function (cosine)

# Plot the data
plt.plot(x, y1, label='Sine', color='blue')
plt.plot(x, y2, label='Cosine', color='red')

# Add labels, title, and legend
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Sine and Cosine Functions')
plt.gcf().set_size_inches(10, 5)
plt.legend()

# Display the plot
plt.show()
```

```python
# Plotting a line chart of prices over time.
prices.plot(kind='line', figsize=(14, 7))
plt.title(symbol + ' Prices Chart')
plt.xlabel('Time')
plt.ylabel('Price')
plt.grid(True)  # Adding a grid for better readability.
```

```python
# Dropping unnecessary columns from the DataFrame.
subset_prices = prices.drop(['Volume', 'Open', 'High', 'Low', 'Adj Close'], axis=1)
subset_prices.head(10)
```

```python
# Plotting the subset of prices with specified formatting.
subset_prices.plot(kind='line', figsize=(14, 7), color='r')
plt.title(symbol + ' Price Chart')
plt.grid(True)
plt.xlabel('Date')
plt.ylabel('Price')
plt.gca().set_facecolor('#d4edda')
```

```python
# Calculating and printing mean and standard deviation of the subset prices.
print('MEAN ', subset_prices.mean())
print('STANDARD DEVIATION', subset_prices.std())
```

```python
# Printing summary statistics for the subset prices.
print('SUMMARY STATISTICS ')
subset_prices.describe()
```

```python
# Modifying prices by multiplying them by 2 and subtracting 10.
modified_prices = subset_prices * 2 - 10
modified_prices.head(5)
```

```python
# Adding noise to prices using a random normal distribution.
noisy_prices = subset_prices + 5 * pd.Series(np.random.normal(0, 5, len(subset_prices)), index=prices.index) + 20
noisy_prices.head(5)
```

```python
# Calculating absolute returns and percentage returns.
add_return = subset_prices.diff()[1:]
mult_return = subset_prices.pct_change()[1:]
```

```python
# Plotting multiplicative returns.
mult_return.plot(kind='line', color='y', figsize=(14, 7))
plt.grid(True)
plt.title('Multiplicative return of ' + symbol)
plt.xlabel('Date')
plt.ylabel('Price')
plt.gca().set_facecolor('#f5f5dc')
```

```python
# Creating a new DataFrame 's' with adjusted prices and additional features.
s = prices.drop(['Volume', 'Open', 'High', 'Low', 'Adj Close'], axis=1)

# Plotting the modified DataFrame 's'.
s.plot(kind='line', figsize=(14, 7), colormap='hot')
plt.title(symbol + ' Price Chart')
plt.grid(True)
plt.xlabel('Date')
plt.ylabel('Price')
plt.gca().set_facecolor('#d4edda')
```

```python
# Calculating and plotting the 30-day rolling mean.
rolling_mean = s.rolling(window=30).mean()
rolling_mean.name = "30 Day Rolling Mean"
```

```python
import matplotlib.pyplot as plt

# Create a figure with two subplots (1 row, 2 columns)
fig, axs = plt.subplots(1, 2, figsize=(14, 7))

# First subplot (left)
s.plot(kind='line', color='r', ax=axs[0])
axs[0].set_title(symbol + ' Price Chart')
axs[0].grid(True)
axs[0].set_xlabel('Date')
axs[0].set_ylabel('Price')
axs[0].set_facecolor('#d4edda')

# Second subplot (right)
rolling_mean.plot(kind='line', color='g', ax=axs[1])
axs[1].set_title('Moving Average chart ' + symbol)
axs[1].set_xlabel('Date')
axs[1].set_ylabel('Price')

# Adjust spacing between subplots
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Calculating the 30-day rolling standard deviation for the DataFrame 's'.
rolling_std = s.rolling(window=30).std()
rolling_std.name = '30 Day Rolling Standard Deviation'
```

```python
# Plotting the 30-day rolling standard deviation.
rolling_std.plot(kind='line', color='m', figsize=(14, 7))
plt.grid(True)
plt.title('30 Day Rolling Standard Deviation of ' + symbol)
plt.xlabel('Date')
plt.ylabel('Standard Deviation')
plt.gca().set_facecolor('#f5f5dc')
```

```python
# Plotting the original price data and the 30-day rolling mean.
x = s.index
plt.figure(figsize=(14, 5))
plt.plot(x, s, label='Price data', color='b')
plt.plot(x, rolling_mean, label='Mean Data', color='r')

plt.title('Price and Mean graph of ' + symbol)
plt.xlabel('Time')
plt.ylabel('Price')
plt.grid(True)
plt.legend()
plt.show()
```
